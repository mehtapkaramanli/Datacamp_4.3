```python
for i in range(0, len(states)):
    if pd.isnull(states.OTHER_EXPENDITURE[i]):
        if (pd.notnull(states.TOTAL_EXPENDITURE[i])and pd.notnull(states.INSTRUCTION_EXPENDITURE[i])and pd.notnull(states.SUPPORT_SERVICES_EXPENDITURE[i])):
            states.OTHER_EXPENDITURE[i]=states.TOTAL_EXPENDITURE[i]-states.INSTRUCTION_EXPENDITURE[i]-states.SUPPORT_SERVICES_EXPENDITURE[i]
```

```python
count1=0
for i in range(0, len(states)):
    if (pd.notnull(states.CAPITAL_OUTLAY_EXPENDITURE[i]) and pd.notnull(states.OTHER_EXPENDITURE[i]) and pd.notnull(states.TOTAL_EXPENDITURE[i])and pd.notnull(states.INSTRUCTION_EXPENDITURE[i])and pd.notnull(states.SUPPORT_SERVICES_EXPENDITURE[i])):
        count1+=1
count1
#count1 sayısı df.info dakıyle aynı, yanı matematıksel bos alanımız yok
```
