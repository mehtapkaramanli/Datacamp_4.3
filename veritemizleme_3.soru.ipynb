```python
import pandas as pd
import numpy as np
import math
states = pd.read_csv('states_all.csv')
```

Öncelikle datasetindeki matematiksel boşlukları aşağıdaki formullerle dolduruyoruz.

    states.GRADES_ALL=states.GRADES_PK+states.GRADES_1_8+states.GRADES_9_12
    states.TOTAL_REVENUE=states.FEDERAL_REVENUE+states.STATE_REVENUE+states.LOCAL_REVENUE
    states.TOTAL_EXPENDITURE=states.INSTRUCTION_EXPENDITURE+states.SUPPORT_SERVICES_EXPENDITURE+states.OTHER_EXPENDITURE+states.CAPITAL_OUTLAY_EXPENDITURE

```python
for i in range(0, len(states)):
    if pd.isnull(states.OTHER_EXPENDITURE[i]):
        if (pd.notnull(states.TOTAL_EXPENDITURE[i])and pd.notnull(states.INSTRUCTION_EXPENDITURE[i])and pd.notnull(states.SUPPORT_SERVICES_EXPENDITURE[i])):
            states.OTHER_EXPENDITURE[i]=states.TOTAL_EXPENDITURE[i]-states.INSTRUCTION_EXPENDITURE[i]-states.SUPPORT_SERVICES_EXPENDITURE[i]
```

```python
count2=0
for i in range(0, len(states)):
    if (pd.notnull(states.TOTAL_REVENUE[i]) and pd.notnull(states.FEDERAL_REVENUE[i])and pd.notnull(states.STATE_REVENUE[i])and pd.notnull(states.LOCAL_REVENUE[i])):
        count2+=1
count2
```

```python
for i in range(0, len(states)):
    if pd.isnull(states.GRADES_ALL_G[i]):
        if (pd.notnull(states.GRADES_9_12_G[i])and pd.notnull(states.GRADES_1_8_G[i])and pd.notnull(states.GRADES_PK_G[i])):
            states.GRADES_ALL_G[i]=states.GRADES_9_12_G[i]+states.GRADES_1_8_G[i]+states.GRADES_PK_G[i]
```
