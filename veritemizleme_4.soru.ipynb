```python
for i in range(0, len(states4)):
    if pd.isnull(states4.OTHER_EXPENDITURE[i]):
        if (pd.notnull(states4.TOTAL_EXPENDITURE[i])and pd.notnull(states4.INSTRUCTION_EXPENDITURE[i])and pd.notnull(states4.SUPPORT_SERVICES_EXPENDITURE[i])):
            states4.OTHER_EXPENDITURE[i]=states4.TOTAL_EXPENDITURE[i]-states4.INSTRUCTION_EXPENDITURE[i]-states4.SUPPORT_SERVICES_EXPENDITURE[i]
```

```python
count2=0
for i in range(0, len(states4)):
    if (pd.notnull(states4.TOTAL_REVENUE[i]) and pd.notnull(states4.FEDERAL_REVENUE[i])and pd.notnull(states4.STATE_REVENUE[i])and pd.notnull(states4.LOCAL_REVENUE[i])):
        count2+=1
count2
```

```python
for i in range(0, len(states4)):
    if pd.isnull(states4.GRADES_ALL_G[i]):
        if (pd.notnull(states4.GRADES_9_12_G[i])and pd.notnull(states4.GRADES_1_8_G[i])and pd.notnull(states4.GRADES_PK_G[i])):
            states4.GRADES_ALL_G[i]=states4.GRADES_9_12_G[i]+states4.GRADES_1_8_G[i]+states4.GRADES_PK_G[i]
```
